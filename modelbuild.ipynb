```python
yaml_content = """# data.yaml
path: /content/helmet and accident
train: train/images
val: valid/images
test: test/images
nc: 3
names: ["With Helmet", "Without Helmet", "Accident"]
"""

with open("/content/helmet and accident/data.yaml", "w") as f:
    f.write(yaml_content)
```

```python
from ultralytics import YOLO

# Load pre-trained YOLOv8 model
model = YOLO("yolov8m.pt")

# Train with augmentation
model.train(
    data="/content/helmet and accident/data.yaml",  # path to your YAML file
    epochs=50,
    imgsz=640,
    # Augmentation settings
    augment=True,          # enables default augmentation
    mosaic=True,           # enables mosaic augmentation
    mixup=True,            # enables mixup augmentation
    hsv_h=0.015,           # hue change
    hsv_s=0.7,             # saturation change
    hsv_v=0.4,             # value change
    degrees=0.0,           # image rotation
    translate=0.1,         # image translation
    scale=0.5,             # image scaling
    shear=0.0,             # shear
    perspective=0.0,       # perspective transform
    flipud=0.0,            # vertical flip
    fliplr=0.5             # horizontal flip
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs/detect/train/weights/best.pt")
metrics = model.val()
print(metrics)
```

```python
# Run validation
results = model.val()

# Overall metrics
print("📊 Overall Metrics:")
print(results)

# Per-class metrics
print("\n📊 Per-Class Metrics:")
for class_id, class_name in results.names.items():
    precision = results.box.p[class_id]
    recall = results.box.r[class_id]
    map50 = results.box.maps[class_id]   # mAP@0.5
    map5095 = results.box.maps[class_id] # (for full IoU 0.5:0.95 if needed)
    print(f"{class_name} -> Precision: {precision:.3f}, Recall: {recall:.3f}, mAP50: {map50:.3f}")
```
